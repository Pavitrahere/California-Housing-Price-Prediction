- The graph is not normally distributed.

```python
house['price_Log'] = np.log(house['price'])

sns.distplot(house['price_Log']);

print("Skewness: %f" % house['price_Log'].skew())
print("Kurtosis: %f" % house['price_Log'].kurt())

house.drop('price', axis= 1, inplace=True)
```

**DISCUSSION AND RESULT**

```python
x = ['area','bedrooms','bathrooms','stories','mainroad','guestroom','airconditioning','parking','prefarea','unfurnished','areaperbedroom']
y = ['price_Log']
```

```python
from sklearn.model_selection import train_test_split

SEED = 42
x_train, x_test, y_train, y_test = train_test_split(house[x],house[y], test_size=0.25, random_state=SEED)
```

1. Linear Regression

**Problem statement
This dataset provides Housing information which includes many attributes such as Area, Stories, Rooms. On the other hand,
the dependant/target variable of the dataset will be Price. So, the problem statement will be to predict the Price of the house
using the independant variables given. The target variable output will be a continuous data, House Price.**

**Suitable Machine Learning Model
Linear Regression model will be the suitable model for the problem statement. Linear regression will be one of the important
supervised learning algorithm used to predict the House Prices. As a result, linear regression is used for predictive modelling
of continuous data rather than categorization. We used Linear Regression to determine to what extent Area or Parking lot impacts
the House Price. Linear regression maps a constant slope using the input, Independant variables with output, Dependant variables
to predict continuous value. It creates a line of best fit that is measured from a scatter plot of data points.**


```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()

regression.fit(x_train,y_train)
```

```python
print(regression.score(x_test,y_test))
```

```python
y_pred = regression.predict(x_test)
```

```python
df_prediction = pd.DataFrame({'Actual' : y_test.squeeze(), 'Predicted': y_pred.squeeze()})
print(df_prediction)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print('mae', mae)
print('mean squared error', mse)
print('R2 Score', r2)
```

2.  Polynomial Regression

**Problem statement
This dataset provides Housing information which includes many attributes such as Area, Stories, Rooms. On the other hand,
the dependant/target variable of the dataset will be Price. So, the problem statement will be to predict the Price of the house
using the independant variables given. The target variable output will be a continuous data, House Price.**

**Suitable Machine Learning Model
Polynomial regression model will be also a alternative model to support the problem statement. So, this algorithm creates polynomial features even before creating a best line fit of regression. This algorithm help us to develop feature space mapping along with linear regression. Moreover, we are using polynomial regression for a better accuracy compare to linear regression because we realise that the dependent and independent variable from this housing dataset can't be best describe with the straight line. As we conclude polynomial regression give us the better accuracy compare to linear regression because the curve line from polynomial provides better prediction for the housing data.** 



```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree = 2, include_bias = False)
poly_features = poly.fit_transform(house[x])
x_train, x_test, y_train, y_test = train_test_split(poly_features,house[y], test_size=0.25, random_state=SEED)
```

```python
poly_reg_model = LinearRegression()
poly_reg_model.fit(x_train, y_train)
```

```python
print(poly_reg_model.score(x_train,y_train))
```

```python
poly_y_predict = poly_reg_model.predict(x_test)
```

```python
df_prediction_poly = pd.DataFrame({'Actual' : y_test.squeeze(), 'Predicted': poly_y_predict.squeeze()})
print(df_prediction_poly)
```
